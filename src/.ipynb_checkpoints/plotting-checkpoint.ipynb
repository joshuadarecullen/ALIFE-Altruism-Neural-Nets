## 3D plots, varying population sizes and mutation rates.

```python
def load_seed_data(seeds):
    all_seed_data = []
    for seed in seeds:
        all_seed_data.append(load_data(f'/home/joshua/Documents/AL/Project/altruism_model/data/seed{seed}-data.txt'))
    return all_seed_data 
```

```python
def iterate_seeds(gen: int, data: dict, key: str):
    
    all_fit = []
    for i, seed_data in enumerate(data):
        if i != 0:
            all_fit += process_altruism(seed_data, key)
        else:
            all_fit = process_altruism(seed_data, key)
        
    return all_fit
        
def process_fittest(data: list, key:str):
    ind_fittest = [[run[gen]['fittest'] for run in mut['data'] ] for mut in data]
    ind_mean = np.array([[sum([indi[key] for indi in best])/20 for best in mut] for mut in ind_fittest])
    ind_mean_runs = np.mean(ind_mean, axis=1).reshape((5,6))
    return ind_mean_runs
```

```python
mutation_rates = [0.0001, 0.001, 0.01, 0.1, 0.2, 0.3]
pop_sizes = np.arange(100, 505, 100)
```

```python
seed_data = load_seed_data(range(0,3))
all_d = iterate_seeds(gen=49, data=seed_data, key='sis')
del seed_data
seed_data = load_seed_data(range(3,5))
all_nd = iterate_seeds(gen=49, data=seed_data, key='sis')
del seed_data
combined = all_d+all_nd
```

```python
def plot_mut_pop_1(data, mut_rates, pop_sizes, z_label, title=None, path=None):
                                 
    font = {'family': 'serif',   
        'color':  'darkred',    
        'weight': 'normal',     
        'size': 15,             
        }                       
                                 
    fig = plt.figure(figsize=(10,10))
    ax = plt.axes(projection='3d')
                                 
    ax.set_xlabel('Mutation Rates', fontdict=font, labelpad=15)
    ax.set_ylabel('Population Size', fontdict=font, labelpad=15)
    ax.set_zlabel(z_label, fontdict=font, labelpad=5)
                                 
     # set up the parameter data for 3d plotting
    X, Y =  np.meshgrid(mut_rates, pop_sizes)
    surf = ax.plot_surface(X, Y, data, cmap=cm.coolwarm, linewidth=0, antialiased=False)
    fig.colorbar(surf, shrink=0.5, aspect=20, label=z_label)
                                 
    if title:                   
        ax.set_title(title, fontdict={'fontsize': 20,
            'fontweight': 20,   
            'color': 'black'})  
                                 
    if path:
        plt.savefig(path+'.png')   
```

## Altruism over generations

```python
def iterate_gens(data: list):
    all_alt_seed = []
    all_alt_not_seed = []
    for i, seed_data in enumerate(data):
        alt_data, alt_not_data = process_gens(seed_data)
        all_alt_seed.append(alt_data)
        all_alt_not_seed.append(alt_not_data)
    return all_alt_seed, all_alt_not_seed
```

```python
def process_gens(data: list):
    alt_seed = []
    alt_not_seed = []
    for i in data:
        alt_run = []
        alt_not_run = []
        for run in i['data']:
            alt_gen = []
            alt_not_gen = []
            for gen in run:
                alt_gen.append(sum([indi['sis'] for indi in gen['fittest']])/20)
                alt_not_gen.append(sum([indi['non_sis'] for indi in gen['fittest']])/20)
            alt_run.append(alt_gen)      
            alt_not_run.append(alt_not_gen)      
        alt_seed.append(list(np.mean(alt_run, axis = 0)))
        alt_not_seed.append(list(np.mean(alt_not_run, axis = 0)))
    return alt_seed, alt_not_seed
```

```python
seed_data = load_seed_data(range(0,3))
alt_seed, alt_not_seed = iterate_gens(data=seed_data)
seed_alt_gens = []
seed_alt_not_gen = []
for i, alt_sd in enumerate(alt_seed):
    seed_alt_gens.append(alt_sd[6])
    seed_alt_not_gen.append(alt_not_seed[i][6])
    
del seed_data
seed_data = load_seed_data(range(3,5))
alt_seed, alt_not_seed = iterate_gens(data=seed_data)
for i, alt_sd in enumerate(alt_seed):
    seed_alt_gens.append(alt_sd[6])
    seed_alt_not_gen.append(alt_not_seed[i][6])
del seed_data  
alt_gen_mean = np.mean(seed_alt_gens, axis = 0)/100
alt_not_gen_mean = np.mean(seed_alt_not_gen, axis = 0)/100
```

```python
title=f'Evolution of Altruism Towards Sister and not a Sister'
def plot_2(A, B, path='/home/joshua/Documents/AL/Project/altruism_model/alt_and_not_alt', title=None):
    font = {'family': 'serif',   
    'color':  'darkred',    
    'weight': 'normal',     
    'size': 15,             
    }         
    plt.plot(A)
    plt.plot(B)  
    plt.legend(['sis','non_sis'])     
    #plt.figure(figsize=(10,10))
    plt.xlabel('Generations')
    plt.ylabel('Amount given')
    if title:                   
        plt.title(title)
    if path:
        plt.savefig(path+'.png')   
```

```python
plot_2(alt_gen_mean, alt_not_gen_mean, title=title)
```

```python
title=f'Evolution of Altruism Towards Sister and not a Sister'
seed_data = load_seed_data(range(0,3))
alt_seed, alt_not_seed = iterate_gens(data=seed_data)
seed_alt_gens = []
seed_alt_not_gen = []
for i, alt_sd in enumerate(alt_seed):
    seed_alt_gens.append(alt_sd[0])
    seed_alt_not_gen.append(alt_not_seed[i][0])
del seed_data
seed_data = load_seed_data(range(3,5))
alt_seed, alt_not_seed = iterate_gens(data=seed_data)
for i, alt_sd in enumerate(alt_seed):
    seed_alt_gens.append(alt_sd[0])
    seed_alt_not_gen.append(alt_not_seed[i][0])
del seed_data  
alt_gen_mean = np.mean(seed_alt_gens, axis = 0)/100
alt_not_gen_mean = np.mean(seed_alt_not_gen, axis = 0)/100
plot_2(alt_gen_mean, alt_not_gen_mean, title=title)
```

## Number of sister groups per generation

```python
def iterate_numblood(data: list):
    numb_seed = []
    for i, seed_data in enumerate(data):
        numb_data = process_bloodgens(seed_data)
        numb_seed.append(numb_data)
    return numb_seed
```

```python
def process_bloodgens(data: list):
    numb_seed = []
    for i in data:
        numb_run = []
        for run in i['data']:
            numb_gen = []
            for gen in run:
                numb_gen.append(gen['num_bloodlines'])
            numb_run.append(numb_gen)      
        numb_seed.append(list(np.mean(numb_run, axis = 0)))
    return numb_seed
```

```python
title=f'Number of bloodlines over Generations'
seed_data = load_seed_data(range(0,3))
numbs = iterate_numblood(data=seed_data)
numb_gens = []
for i, numb in enumerate(numbs):
    numb_gens.append(numb[6])
del seed_data

seed_data = load_seed_data(range(3,5))
numbs = iterate_numblood(data=seed_data)
for i, numb in enumerate(numbs):
    numb_gens.append(numb[6])
del seed_data  

numb_mean = np.mean(numb_gens, axis = 0)
plt.plot(numb_mean)
plt.xlabel('Generations')
plt.ylabel('Amount of bloodlines')
plt.title(title)
plt.savefig('/home/joshua/Documents/AL/Project/altruism_model/numblood200.png')   
```
